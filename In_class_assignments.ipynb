```python
sales = float(input("Enter sales ($): "))
cogs = sales * 0.4
print(f"Sales: {sales}$")
print(f"Costs of Goods Sold: {cogs}$")
print(f"Gross Profit: {sales - cogs}$")
```

```python
x = float(input("Enter a number: "))
y = float(input("Enter another number: "))
calc = input("Enter a operator (+,-,/,*,%): ")

if calc == "+":
  print(x + y)
elif calc == "-":
  print(x - y)
elif calc == "/":
  print(x / y)
elif calc == "*":
  print(x * y)
elif calc == "%":
  print(x % y)
else:
  print("Invalid Operator")

```
